```python
"""
Création de la base de données mmo-projet-kayak
"""
import os
from dotenv import load_dotenv
import psycopg2
from psycopg2 import sql

load_dotenv()

host = os.getenv('DB_HOST_2')
port = int(os.getenv('DB_PORT_2', 5432))
user = os.getenv('DB_USER_2')  # postgres
password = os.getenv('DB_PASSWORD_2')

print("🔗 Connexion à la database 'postgres' (défaut)...")

try:
    # Connexion à la database par défaut 'postgres'
    connection = psycopg2.connect(
        host=host,
        port=port,
        user=user,
        password=password,
        database='postgres',  # Base par défaut
        connect_timeout=10
    )
    
    # IMPORTANT : autocommit pour CREATE DATABASE
    connection.autocommit = True
    
    print("✅ Connecté !")
    
    with connection.cursor() as cursor:
        # Vérifie si la database existe déjà
        cursor.execute(
            "SELECT 1 FROM pg_database WHERE datname = %s",
            ('mmo-projet-kayak',)
        )
        exists = cursor.fetchone()
        
        if exists:
            print("ℹ️  La database 'mmo-projet-kayak' existe déjà.")
        else:
            print("📝 Création de la database 'mmo-projet-kayak'...")
            
            # Crée la database avec un nom entre guillemets à cause du tiret
            cursor.execute(
                sql.SQL('CREATE DATABASE {}').format(
                    sql.Identifier('mmo-projet-kayak')
                )
            )
            
            print("✅ Database 'mmo-projet-kayak' créée avec succès !")
    
    connection.close()
    print("\n🎉 Tu peux maintenant te connecter à 'mmo-projet-kayak' !")
    
except psycopg2.Error as e:
    print(f"❌ Erreur : {e}")
```

```python
"""
Test de connexion TCP uniquement
"""
import socket

host = os.getenv('DB_HOST_2')
port = 5432

print(f"🔍 Test de connexion TCP vers {host}:{port}")

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
sock.settimeout(10)

try:
    sock.connect((host, port))
    print("✅ CONNEXION TCP RÉUSSIE !")
    print("💡 Le security group fonctionne correctement.")
  
    sock.close()
except Exception as e:
    print(f"❌ Échec de la connexion TCP : {e}")
```

```python
"""
Script de Test de Connexion AWS RDS PostgreSQL
===============================================
Ce script teste la connexion à une base de données RDS PostgreSQL.
"""

import os
from dotenv import load_dotenv
import psycopg2
from psycopg2 import sql

# Charge les variables d'environnement
load_dotenv()

# Configuration
host = os.getenv('DB_HOST_2')
port = int(os.getenv('DB_PORT_2'))  # Port par défaut PostgreSQL
database = os.getenv('DB_NAME_2')
user = os.getenv('DB_USER_2')
password = os.getenv('DB_PASSWORD_2')
try:
    # Connexion PostgreSQL
    connection = psycopg2.connect(
        host=host,
        port=port,
        user=user,
        password=password,
        database=database,
        connect_timeout=10
    )
    
    print("✅ Connexion réussie à la base de données RDS PostgreSQL!")
    
    # Test simple
    with connection.cursor() as cursor:
        # Version PostgreSQL
        cursor.execute("SELECT VERSION()")
        version = cursor.fetchone()
        print(f"📊 Version PostgreSQL : {version[0]}")
        
        # Utilisateur courant
        cursor.execute("SELECT current_user")
        current_user = cursor.fetchone()
        print(f"👤 Utilisateur : {current_user[0]}")
        
        # Base de données courante
        cursor.execute("SELECT current_database()")
        current_db = cursor.fetchone()
        print(f"🗄️  Base de données : {current_db[0]}")
    
    connection.close()
    print("✅ Connexion fermée correctement")
    
except psycopg2.Error as e:
    print(f"❌ Erreur de connexion PostgreSQL : {e}")
    print("\n🔍 Vérifications à faire :")
    print("   1. L'instance RDS est-elle en état 'Available' ?")
    print("   2. Le Security Group autorise-t-il ton IP sur le port 5432 ?")
    print("   3. L'endpoint est-il correct ?")
    print("   4. Les identifiants sont-ils corrects ?")
    
except Exception as e:
    print(f"❌ Erreur générale : {e}")


```
