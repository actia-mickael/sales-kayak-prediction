```python
"""
Création de la base de données mmo-projet-kayak sur MySQL
"""
import os
from dotenv import load_dotenv
import mysql.connector
from mysql.connector import Error

load_dotenv()

host = os.getenv('DB_HOST_3')
port = int(os.getenv('DB_PORT_3', 3306))
user = os.getenv('DB_USER_3')
password = os.getenv('DB_PASSWORD_3')

print("🔗 Connexion au serveur MySQL...")

try:
    # Connexion au serveur MySQL (sans spécifier de database)
    connection = mysql.connector.connect(
        host=host,
        port=port,
        user=user,
        password=password,
        connect_timeout=10
    )
    
    print("✅ Connecté au serveur MySQL !")
    
    cursor = connection.cursor()
    
    # Vérifie si la database existe déjà
    cursor.execute("SHOW DATABASES LIKE 'mmo_projet_kayak'")
    exists = cursor.fetchone()
    
    if exists:
        print("ℹ️  La database 'mmo_projet_kayak' existe déjà.")
    else:
        print("📝 Création de la database 'mmo_projet_kayak'...")
        
        # Note: MySQL n'aime pas les tirets dans les noms de database
        # On utilise des underscores à la place
        cursor.execute("CREATE DATABASE `mmo_projet_kayak` CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci")
        
        print("✅ Database 'mmo_projet_kayak' créée avec succès !")
    
    # Liste toutes les databases
    print("\n📚 Databases disponibles sur le serveur :")
    cursor.execute("SHOW DATABASES")
    for db in cursor.fetchall():
        print(f"   • {db[0]}")
    
    cursor.close()
    connection.close()
    print("\n🎉 Tu peux maintenant te connecter à 'mmo_projet_kayak' !")
    
except Error as e:
    print(f"❌ Erreur MySQL : {e}")
```
