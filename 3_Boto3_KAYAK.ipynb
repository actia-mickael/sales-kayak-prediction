### sauvegarde data dans mon S3

```python
import boto3
import os

#acces 1 
YOUR_ACCESS_KEY = os.getenv("AWS_ACCESS_KEY_ID")
YOUR_SECRET_KEY = os.getenv("AWS_SECRET_ACCESS_KEY")

session = boto3.Session(aws_access_key_id=YOUR_ACCESS_KEY, 
                        aws_secret_access_key=YOUR_SECRET_KEY)
```

```python
Nom_Bucket = "dossier-kayak-jedha-fullstack"    
#essai de connexion
# création du bucket
s3 = session.resource("s3")
# Tentative de listage des buckets (c'est l'opération de test)
print("Mes buckets :")
for bucket in s3.buckets.all():
    print(f" - {bucket.name}")
    
Name_bucket =  Nom_Bucket
if Name_bucket in [bucket.name for bucket in s3.buckets.all()]:
    print(F"Nota: Le bucket {Name_bucket}est deja existant.")
else:
    bucket = s3.create_bucket(
        Bucket=Name_bucket,
        CreateBucketConfiguration={'LocationConstraint': 'eu-north-1' },)
```

```python
# Upload des fichiers sur S3
with open("booking_hotels_france5.csv", "rb") as data_hotels:
    bucket.put_object(Key="booking_hotels_france5.csv", Body=data_hotels)

with open("Meteo_villes_france.csv", "rb") as data_meteo:
    bucket.put_object(Key="Meteo_villes_france.csv", Body=data_meteo)

with open("Meteo_villes_france2.csv", "rb") as data_meteo2:
    bucket.put_object(Key="Meteo_villes_france2.csv", Body=data_meteo2)

with open("booking_hotels_france_with_address.csv", "rb") as data_hotels_adr:
    bucket.put_object(Key="booking_hotels_france_with_address.csv", Body=data_hotels_adr)

print("Les fichiers ont été uploadés avec succès dans le bucket.")
```

```python
bucket = s3.Bucket(Name_bucket)
print(f"--- Fichiers dans {bucket.name} ---")

for obj in bucket.objects.all():
    print(obj.key)
```
