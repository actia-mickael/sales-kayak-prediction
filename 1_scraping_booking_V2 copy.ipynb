```python
# Liste pour stocker tous les DataFrames
all_dfs = []

for ville in villes:
    print(f"Scraping {ville}...")
    
    params = {"ss" : ville,
          "nflt": type_logement,
          "checkin" : date_debut,
          "checkout" : date_fin,
          "group_adultes":2,
          "group_children" :0}
    response = requests.get(url=url_base, headers=headers, params=params)
    
    if response.status_code != 200:
        print(f"  ⚠ Erreur {response.status_code} pour {ville}")
        continue
    
    selector = Selector(text=response.text)
    hotels = selector.css("div[class='aa97d6032f']")
    
    # Les selectors
    name_hotels = hotels.css("div[class='b87c397a13 a3e0b4ffd1']::text").getall()
    notes_hotels = hotels.css("div[class='f63b14ab7a dff2e52086']::text").getall()
    web_hotels = hotels.css('a[data-testid="title-link"]::attr(href)').getall()
    experience_hotels = hotels.css("div[class='fff1944c52 fb14de7f14 eaa8455879']::text").getall()
    descri_hotels = hotels.css("h4[class='fff1944c52 f254df5361']::text").getall()
     
    # Création de la map du DataFrame pour cette ville
    n = len(name_hotels)
    df_ville = pd.DataFrame({
        "ville": [ville] * n,
        "nom": name_hotels,
        "note": notes_hotels[:n],  # Ajuster la longueur si nécessaire
        "site_web": web_hotels[:n],
        "nb_experience": experience_hotels[:n],
        "description": descri_hotels[:n],
    })
    
    all_dfs.append(df_ville)
    print(f"  ✓ {len(df_ville)} hôtels trouvés")
    
    # Pause pour éviter le blocage
    time.sleep(2)

# Concaténer tous les DataFrames
df_final = pd.concat(all_dfs, ignore_index=True)
df_final.index = df_final.index + 1

# Trier par note
df_final["note"] = pd.to_numeric(df_final["note"].str.replace(",", "."), errors="coerce")
df_final = df_final.sort_values(by="note", ascending=False)

# Sauvegarder
df_final.to_csv("booking_hotels_france5.csv", index=True)
print(f"\n✓ Total : {len(df_final)} hôtels sauvegardés")
```
